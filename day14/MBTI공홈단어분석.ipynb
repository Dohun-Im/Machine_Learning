```python
import nltk
import pandas as pd
```

```python
file = open('INFP.txt', encoding='utf8')
infp = file.read()
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tag import pos_tag, untag
from nltk import Text
from nltk import FreqDist
from nltk.tag import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
```

```python
retokenize = RegexpTokenizer('[\w]+')

text = Text(retokenize.tokenize(infp))
text
```

```python
text.plot(15)
```

```python
from nltk import FreqDist
from nltk.tag import pos_tag

stopwords = list(line.strip() for line in open('stopwords.txt'))

infp_tokens = pos_tag(retokenize.tokenize(infp))
```

```python

names_list_infp = [t[0] for t in infp_tokens if t[0] not in stopwords] #t[1] =='NN' and
for index, value in enumerate(names_list_infp):
    if value == 'Mediators':
        names_list_infp[index] = 'Mediator'


fd_names_infp = FreqDist(names_list_infp)
```

```python
fd_names_infp.N()
```

```python
fd_names_infp.plot(10)
```

```python
fd_names_infp
```

```python
fd_names_infp.freq('good')
```

```python
print(fd_names_infp.most_common(10))
```

```python
lst = fd_names_infp.most_common(10)
df = pd.DataFrame(lst, columns = ['Word','Count'])
```

```python
import matplotlib.pyplot as plt

df.plot.bar(x='Word', y='Count', figsize=(20,10))
plt.show()
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600,
               background_color='white',
              random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names_infp))
plt.show()
```
