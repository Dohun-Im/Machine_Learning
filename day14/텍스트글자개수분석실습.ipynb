```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

file1 = open('pos.txt', encoding='utf8')
file2 = open('neg.txt', encoding='utf8')
a = file1.read()
b = file2.read()
```

```python
for line in open('stopwords.txt'):
    print(line.strip())
```

```python
stopwords = set(line.strip() for line in open('stopwords.txt')) #익명함수, 클러져
stopwords 
```

```python
stopwords = stopwords.union(set(['(',')','2021','and','the','to''mr','mrs','one','two'])) #union = 합집합
stopwords  
```

```python
wordcount = {}
for word in a.lower().split():
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("''\''","")
    word = word.replace("*","")
    
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1 
```

```python
wordcount
```

```python
import collections
```

```python
word_counter = collections.Counter(wordcount)
```

```python
word_counter.most_common(3)
```

```python
n_print = int(input('how many list>> '))
for word, count in word_counter.most_common(n_print):
    print(word, ':', count)
```

```python
file1.close()
```

```python
lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word','Count'])
df
```

```python
df.plot.bar(x='Word',y='Count')
```
