```python
import pandas as pd
import glob
import re
from functools import reduce

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
all_files = glob.glob("./data_list/my*.xls")
all_files
```

```python
all_files_data = []
```

```python
for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)
    
all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index=True)
all_files_data_concat
```

```python
all_title = all_files_data_concat['제목']
all_title
```

```python
type(all_title)
```

```python
# iterator data(반복자) - list, Series -> 반복형태의 데이터
```

```python
stopWords= set(stopwords.words("english"))
lemma = WordNetLemmatizer()
```

```python
words = []

for title in all_title:
    # 문자가 아닌 것을 공백으로 대체
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title)) 
    
    # 소문자로 변환 후 토큰화(리스트로)
    EnWordsToken = word_tokenize(EnWords.lower())
   
    # 토큰화한 단어들이 stopwords에 있는지 확인. 있으면 제거
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords] #익명함수, 클로져
#    위에 익명함수와 같은것
#     def call(EnWordsToken):
#         result = []
#         for w in EnWordsToken:
#             if w not in stopwords:
#                 result.append(w)
#         return result
    
    
    # 꺼내서 표제어 선정까지
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    
#     def call2(EnWordsTokenStop):
#         result2 = []
#         for w in EnWordsTokenStop:
#             result2.append(lemma.lemmatize(w))
#         return result2
    
    words.append(EnWordsTokenStopLemma)
```

```python
words
```

```python
words2 = list(reduce(lambda x,y: x+y,words))
words2
```

```python
count = Counter(words2)

count
```

```python
word_count = dict()

for tag, counts  in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```
